```python
#Load independencies
import pandas as pd
from datetime import datetime as dt
import os
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
import matplotlib.dates as mdates
import datetime as dt
import numpy as np
```

```python
Teen_NoSchool_Byrace_df = pd.read_excel("NotHSGrad_NoSchool_Year1619_ByRace.xlsx")
Teen_NoSchool_Byrace_df.head(10)
```

```python
#Grabbing Race unqual to Total and only numbers
Teen_NoSchool_Byrace_df = Teen_NoSchool_Byrace_df.loc[(Teen_NoSchool_Byrace_df["DataFormat"] == "Number") & (Teen_NoSchool_Byrace_df["Race"] != "Total") & (Teen_NoSchool_Byrace_df["Race"] != "Two or more races")]
Teen_NoSchool_Byrace_df.head()
```

```python
#Getting rid of NA data
Teen_NoSchool_Byrace_df = Teen_NoSchool_Byrace_df.loc[(Teen_NoSchool_Byrace_df["Data"] != "N.A.")]
Teen_NoSchool_Byrace_df.head(10)
```

```python
Teen_NoSchool_Byrace_df.value_counts("Race")
```

```python
#Setting our timeframe between 2005 and 2013
Teen_NoSchool_Byrace_df = Teen_NoSchool_Byrace_df.loc[(Teen_NoSchool_Byrace_df["TimeFrame"] >= 2005) & (Teen_NoSchool_Byrace_df["TimeFrame"] < 2014)]
Teen_NoSchool_Byrace_df.head()
```
