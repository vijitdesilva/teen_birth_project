```python
#Load independencies
import pandas as pd
from datetime import datetime as dt
import os
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
import matplotlib.dates as mdates
import datetime as dt
import numpy as np
```

```python
#Reading Files
US_city_info = pd.read_csv('uscities.csv')
US_city_info.head()
```

```python
US_city_info = US_city_info.loc[(US_city_info['state_name'] != 'District of Columbia')]
US_city_info = US_city_info.loc[:, ('state_id', 'state_name', 'population', 'lat', 'lng')]
US_city_info.head()
```

```python
#US_city_info = US_city_info.loc[(US_city_info('state_name') != 'District of Columbia')]
#US_city_info.head()
```

```python
US_city_info.rename(columns =  {'state_id': 'State'}, inplace = True)
```

```python
US_city_info.rename(columns =  {'state_name': 'Location'}, inplace = True)
```

```python
US_city_info.rename(columns =  {'population': 'Population'}, inplace = True)
```

```python
US_city_info.rename(columns =  {'lat': 'Latitude'}, inplace = True)
```

```python
US_city_info.rename(columns =  {'lng':'Longitude'}, inplace = True)
US_city_info.head()
```

```python
US_grouped_city_info = US_city_info.groupby(['Location'])
```

```python
US_grouped_city_info.head(25)
```

```python
US_latitude_state_info = US_grouped_city_info['Latitude'].mean()
US_longitude_state_info = US_grouped_city_info['Longitude'].mean()
US_population_state_info = US_grouped_city_info['Population'].sum()
US_longitude_state_info.head(50)
```

```python
US_latitude_state_info.head(50)
```

```python
US_population_state_info.head(50)
```

```python
merged_df = pd.merge(US_latitude_state_info, US_longitude_state_info, on='Location', how='left')
merged_df.head()
```

```python
merged_df = pd.merge(merged_df, US_population_state_info, on='Location', how='left')
merged_df.head()
```

```python
merged_df.to_csv('State_Coord_Population.csv', index=False, header=True)
```
