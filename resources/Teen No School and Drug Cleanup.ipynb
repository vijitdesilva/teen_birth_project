```python
#Reading files
#Teen_Drug_df = pd.read_csv("Alcohol_Drug_Teen.csv")
Teen_No_School_df = pd.read_excel("NotHSGrad_NoSchool_Year1619_ByRace.xlsx")
```

```python
#Grabbing Total and only numbers
Teen_No_School_df= Teen_No_School_df.loc[(Teen_No_School_df["DataFormat"] == "Number") & (Teen_No_School_df["Race"] == "Total")]
Teen_No_School_df.head()
```

```python
#Getting rid of NA data
Teen_No_School_df = Teen_No_School_df.loc[(Teen_No_School_df["Data"] != "N.A.") & (Teen_No_School_df["Data"] != "S")]
Teen_No_School_df
```

```python
#Setting our timeframe between 2005 and 2014
Teen_No_School_df = Teen_No_School_df.loc[(Teen_No_School_df["TimeFrame"] >= 2005) & (Teen_No_School_df["TimeFrame"] < 2014)]
Teen_No_School_df.head()
```

```python
#Not showing Virgin Islands, DC, and Puerto Rico because of bad data
Teen_No_School_df = Teen_No_School_df.loc[(Teen_No_School_df["Location"] != "Virgin Islands") & (Teen_No_School_df["Location"] != "Puerto Rico") & (Teen_No_School_df["Location"] != "District of Columbia") & (Teen_No_School_df["Location"] != "United States")]
Teen_No_School_df
```

```python
#Only displaying Location Data and Timeframe for analysis
Teen_No_School_df = Teen_No_School_df.loc[:, ("Location", "Data", "TimeFrame")]
Teen_No_School_df
```
