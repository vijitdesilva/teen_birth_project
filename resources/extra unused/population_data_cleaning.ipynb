```python
Clean_Population_df = Population_df.loc[(Population_df["Notes"]!= "Total")]
Clean_Population_df.head(35)
```

```python
Clean_Population_df = Clean_Population_df.loc[:, ["Gender", "State", "Age Group", "Year", "Projected Populations"]]
Clean_Population_df.head(50)
```

```python
# Extract data from 2005 to 2014
Clean_Population_df = Clean_Population_df.loc[(Clean_Population_df["Year"] >= 2005.0) & (Clean_Population_df["Year"] <2015.0)]
Clean_Population_df.head(10)
```

```python

Clean_Population_df = Clean_Population_df.groupby(["State"])
Clean_Population_df.head(40) 
```

```python
Clean_Population_df = Clean_Population_df["Projected Populations"].mean()
Clean_Population_df
```
