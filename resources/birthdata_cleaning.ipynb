```python
# Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import matplotlib
matplotlib.style.use('ggplot')

import scipy.stats as st
from scipy.stats import linregress
from scipy.stats import sem

import seaborn as sns
```

# Upload data files and read 

```python
Teen_Pregnancy_df = pd.read_excel("Total_teen_births.xlsx")
Teen_Pregnancy_df.head()
```

```python
# Extract Number values from DataFormat
Teen_Pregnancy_df = Teen_Pregnancy_df.loc[(Teen_Pregnancy_df["DataFormat"]== "Number")]
Teen_Pregnancy_df.head()
```

```python
# Extract data from 2005 to 2014
Teen_Pregnancy_df = Teen_Pregnancy_df.loc[(Teen_Pregnancy_df["TimeFrame"] >= 2005) & (Teen_Pregnancy_df["TimeFrame"] <2014)]
Teen_Pregnancy_df.head(10)
```

```python
# Drop any null values
Teen_Pregnancy_df  = Teen_Pregnancy_df.dropna()
Teen_Pregnancy_df.head(10)
```

```python
# Exclude not applicable data 
NA_Data_Df = Teen_Pregnancy_df.loc[(Teen_Pregnancy_df["Data"] == "N.A.") | (Teen_Pregnancy_df["Data"] == "S")]
NA_Data_Df.head(10)
```

```python
# Exclude Virgin Islan due to N.A data 
Teen_Pregnancy_df = Teen_Pregnancy_df.loc[(Teen_Pregnancy_df["Location"] != "Virgin Islands")]
Teen_Pregnancy_df.head(25)
```
